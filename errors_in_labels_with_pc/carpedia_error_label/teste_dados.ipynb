```python
import numpy as np
import matplotlib.pyplot as plt 
from ocpc_py import MultiClassPC
import pandas as pd
import os
```

```python
print(os.listdir())
df_dados_carpedia = pd.read_csv("dados.csv", sep=";")

# Extraindo dados do df
X = df_dados_carpedia.iloc[:, :-1]
Y = df_dados_carpedia.iloc[:,-1]

# Criar gráficos scatter para cada coluna de X
num_cols = X.shape[1]  # Número de colunas em X
fig, axes = plt.subplots(1, num_cols, figsize=(5 * num_cols, 5))

# Se houver apenas uma coluna em X, transformar axes em lista para evitar erro
if num_cols == 1:
    axes = [axes]

for i, col in enumerate(X.columns):
    axes[i].scatter(X[col], Y, alpha=0.7)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Y")
    axes[i].set_title(f"Scatter plot de {col} vs Y")

plt.tight_layout()
plt.show()
```
